```python
import pandas as pd
```

```python
df = pd.read_csv("data/AID_624202_datatable_all.csv")
```

```python
df = df.iloc[5:, :]
df
```

```python
df = df.dropna(subset=["Efficacy"])
df
```

```python
import numpy as np
df['Efficacy'] = df['Efficacy'].astype(float)
df['pchembl'] = -1 * np.log10(df['Efficacy'].astype(float)*10**(-6))
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
import seaborn as sns

# Filter the data by class
data_class1 = df[df['PUBCHEM_ACTIVITY_OUTCOME'] == "Active"]['Efficacy']
data_class2 = df[df['PUBCHEM_ACTIVITY_OUTCOME'] == "Inactive"]['Efficacy']

# Creating a DataFrame suitable for seaborn's violin plot
violin_data = pd.DataFrame({
    'Efficacy': pd.concat([data_class1, data_class2]),
    'Activity': ['Active'] * len(data_class1) + ['Inactive'] * len(data_class2)
})

# Plotting with seaborn
plt.figure(figsize=(10, 6))
sns.violinplot(x='Activity', y='Efficacy', data=violin_data)
plt.title('Violin Plot of Efficacy by Activity Outcome')
plt.ylabel('Efficacy')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
import seaborn as sns

# Filter the data by class
data_class1 = df[df['PUBCHEM_ACTIVITY_OUTCOME'] == "Active"]['pchembl']
data_class2 = df[df['PUBCHEM_ACTIVITY_OUTCOME'] == "Inactive"]['pchembl']

# Creating a DataFrame suitable for seaborn's violin plot
violin_data = pd.DataFrame({
    'Efficacy': pd.concat([data_class1, data_class2]),
    'Activity': ['Active'] * len(data_class1) + ['Inactive'] * len(data_class2)
})

# Plotting with seaborn
plt.figure(figsize=(10, 6))
sns.violinplot(x='Activity', y='Efficacy', data=violin_data)
plt.title('Violin Plot of Efficacy by Activity Outcome')
plt.ylabel('Efficacy')
plt.show()
```

```python
# Performing Welch's t-test
t_stat, p_value = ttest_ind(data_class1, data_class2, equal_var=False)

print(f"T-statistic: {t_stat}, P-value: {p_value}")
```
