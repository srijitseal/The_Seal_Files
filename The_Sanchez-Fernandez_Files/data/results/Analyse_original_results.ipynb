```python
import pandas as pd
```

```python
df1 = pd.read_csv("cloome-bioactivity_split1.csv")
df1 = df1[df1["Test_AUC"]!=0.500]
df1
```

```python
df2 = pd.read_csv("cloome-bioactivity_split2.csv")
df2 = df2[df2["Test_AUC"]!=0.500]
df2
```

```python
df3 = pd.read_csv("cloome-bioactivity_split3.csv")
df3 = df3[df3["Test_AUC"]!=0.500]
df3
```

```python
df = pd.concat([df1, df2, df3])
df
```

```python
import matplotlib.pyplot as plt

df.N_labels_test.hist()
plt.show()
```

```python
df = df.groupby("Unnamed: 0").mean()

df = df.reset_index().rename(columns={"Unnamed: 0": "Task"})[["Task", "Test_AUC"]]
df
```

```python
df.to_csv("CLOOME_updated_results.csv", index=False)
df
```
