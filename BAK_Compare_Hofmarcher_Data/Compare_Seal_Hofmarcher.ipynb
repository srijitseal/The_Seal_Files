```python
a="CHEMBL1741321"
k = pd.merge(df[["InChIKey",a ]], df2[["InChIKey",a]], left_on="InChIKey", right_on="InChIKey", how="outer")
k = k.replace(np.nan, -1)
k
```

```python
x = f"{a}_x"
y = f"{a}_y"
```

```python
k = k[(k[x] != -1) | (k[y] != -1)]
k
```

```python
k[(k[x] == -1) & (k[y] != -1)]
```

```python
# Assuming your dataframe is named 'df'

print(k[x].value_counts())
    
print(k[y].value_counts())

# Count how many 1's agree
ones_agree = len(k[(k[x] == 1) & (k[y] == 1)])

# Count how many 0's agree
zeros_agree = len(k[(k[x] == 0) & (k[y] == 0)])

# Count how many extra 1's are in 737823_x but not in 737823_y
extra_ones_x = len(k[(k[x] == 1) & (k[y] == -1)])

# Count how many extra 1's are in 737823_y but not in 737823_x
extra_ones_y = len(k[(k[x] == -1) & (k[y] == 1)])

# Count how many extra 0's are in 737823_x but not in 737823_y
extra_zeros_x = len(k[(k[x] == 0) & (k[y] == -1)])

# Count how many extra 0's are in 737823_y but not in 737823_x
extra_zeros_y = len(k[(k[x] == -1) & (k[y] == 0)])

# Count how many conflicts (values that don't match at all)
conflicts = len(k[k[x] != k[y]])

# Print the results
print(f"Number of 1's that agree: {ones_agree}")
print(f"Number of 0's that agree: {zeros_agree}")
print(f"Number of extra 1's in 737823_x: {extra_ones_x}")
print(f"Number of extra 1's in 737823_y: {extra_ones_y}")
print(f"Number of extra 0's in 737823_x: {extra_zeros_x}")
print(f"Number of extra 0's in 737823_y: {extra_zeros_y}")
print(f"Number of conflicts: {conflicts}")
```
