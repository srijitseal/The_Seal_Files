```python
import pandas as pd
```

```python
df = pd.read_csv("label-matrix.mtx", sep = " ")
df
```

```python
compounds = pd.read_csv("row-compound-index.csv", sep = ",")
compounds
```

```python
df = pd.merge(df, compounds, left_on="compoundid", right_on="INDEX")
df
```

```python
assays = pd.read_csv("column-assay-index.csv", sep = ",").reset_index()
assays["index"] = assays["index"]+1
assays
```

```python
assays.index.nunique()
```

```python
df = pd.merge(df, assays, left_on="task", right_on="index")
df
```

```python
# Set the index to 'assay'
df.set_index('ASSAY_ID', inplace=True)

# Create a new DataFrame to ensure 'assay' becomes columns with 'activity' values
df_pivot = df.pivot_table(index='INCHIKEY', columns='ASSAY_ID', values='activity')
df_pivot.reset_index(drop=True)

df_pivot
```

```python
# Resetting the index to make the 'INCHIKEY' a new column
df_pivot.reset_index(inplace=True)
df_pivot.rename(columns={'index': 'INCHIKEY'}, inplace=True)
df_pivot
```

```python
import numpy as np
# Setting inactives to 0, actives to 1, and no data to -1
df_pivot = df_pivot.replace(-1, 0)
df_pivot = df_pivot.replace(np.nan, -1)
df_pivot
```
