```python
df2 = pd.read_csv("data/Unet_Features.csv")
df2
```

```python
df1 = pd.read_csv("data/GroundTruth_Features.csv")
df1
```

```python
import pandas as pd
import numpy as np
from scipy.stats import spearmanr
```

```python
# Merging the dataframes on index (assuming index represents the unique IDs)
merged_df = pd.merge(df1, df2, left_index=True, right_index=True, suffixes=('_df1', '_df2'))

# Dropping rows with missing values (if any)
merged_df_cleaned = merged_df.dropna()
merged_df_cleaned
```

```python
import matplotlib.pyplot as plt
```

```python
spearman_correlations_loop = []

for col in df1.columns[4:]:
    feature_df1 = f"{col}_df1"
    feature_df2 = f"{col}_df2"
    spearman_corr, _ = spearmanr(merged_df_cleaned[feature_df1], merged_df_cleaned[feature_df2])
    spearman_correlations_loop.append((col, spearman_corr))

# Convert results to a DataFrame
spearman_correlations_loop_df = pd.DataFrame(spearman_correlations_loop, columns=["Feature", "Spearman_Correlation"])
spearman_correlations_loop_df["Spearman_Correlation"] = abs(spearman_correlations_loop_df["Spearman_Correlation"])
spearman_correlations_loop_df.hist()
plt.show()
```
